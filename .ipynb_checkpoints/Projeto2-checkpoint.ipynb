O trabalho tinha como objetivo prever a quantidade de turistas que um país receberia. Tendo em vista isso, buscamos utilizar do conhecimento obtido na aula de ciência dos dados para fazer essa análise. Por isso optamos pelo uso de três métodos de regressão, sendo eles regressão múltipla, decision tree e random forest, visto que o resultado buscado seria numérico. 

```python
#Importando bibliotecas

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns; sns.set()
import statsmodels.api as sm
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor
```

### Juntando as bases de dados

Para isso selecionamos primeiro a nossa base de dados. Formulamos a hipótese de que o quanto um país investe em turismo e quanto mais ele recebe em retorno disso, mais turistas ele tem, então buscamos uma base de dados que nos informasse isso. Porém também foi discutido que além desses fatores economicos, o clima e a paisagem do lugar influenciam na decisão, isso foi feito também para enriquecer nosso estudo. Como não havia um dataset que relacionasse os dois, cruzamos as tabela encontradas.

```python
dados_turismo = pd.read_excel('Dados_turismo.xlsx',thousands=",")
dados_turismo = dados_turismo.set_index('Country')
dados_turismo.head()
```

```python
dados_bio = pd.read_excel('Dados_bio.xlsx')
dados_bio = dados_bio.sort_values('Country')
dados_bio.loc[:,'Country'] = dados_bio.loc[:,'Country'].str.strip()
dados_bio = dados_bio.set_index('Country')
dados_bio.head()
```

```python
dados = dados_turismo.merge(dados_bio, left_index=True, right_index=True)
dados.head()
```

```python
dados.describe()
```

### Explorando a correlação entre variáveis

Foi feita uma tabela que nos indicasse a correlação entre as variáveis, para assim determinarmos quais influenciam mais a estimativa do int inbound 2018, que foi nossa variável escolhida.

```python
correlacao = dados.corr()
correlacao.head()
```

```python
#Fizemos isso para checar se todas as variáveis poderiam ser usadas na regressão
dados.dtypes
```

#### Gráfico de correlação entre variáveis
Esse gráfico foi feito para auxiliar na visualização da correlação entre variáveis, isso confirmou a nossa hipótese de alta equiparação entre alguns fatores da base de dados, que é representada pelos quadrados mais claros do heatmap.

```python
grafico = sns.heatmap(correlacao)
```

### Fazendo as Regressões

Primeiramente fizemos uma regressão múltipla utilizando todos as variáveis da base, porém, depois de analisar os resultados verificamos que eles estavam sendo muito influenciados por algumas categorias que como visto previamente, tinham uma correlação muito alta. Então optamos por fazer uma iteração e retirá-los da análise.

```python
#Função que determina os X's e o Y da regressão e da o resultado.

def regress(X,Y):
    '''
    X: coluna(s) do DataFrame que serão utilizadas como variável(is) preditoras/explicativas (independente(s))
    Y: coluna do DataFrame que será utilizada como variável resposta (dependente)
    '''
    X_cp = sm.add_constant(X)
    model = sm.OLS(Y,X_cp)
    results = model.fit()
    return results
```

```python
#dropna tira as linhas que tem o NaN
dados_completos = dados.dropna()

X = dados_completos[['int inbound 2012', 'int outbound 2012',
       'int outbound 2018', 'inbound exp 2012  (millions)',
       'inbound exp 2018  (millions)', 'inbound exp 2012  (percentage)',
       'inbound exp 2018  (percentage)', 'outbound exp 2012  (millions)',
       'outbound exp 2018  (millions)', 'outbound exp 2012  (percentage)',
       'outbound exp 2018  (percentage)', 'Bare area', 'Cropland',
       'Tree cover', 'Grassland', 'Shrubland', 'Sparse vegetation',
       'Artificial surfaces', 'Inland water', 'Wetland']]
Y = dados_completos['int inbound 2018']
```

```python
dados_completos.head()
```

### Regressão Multipla

```python
results = regress(X,Y)
results.summary()
```

#### Aqui é onde retiramos as variáveis que atrapalhavam os resultados, como podemos ver, o P value das variáveis diminuiu, porém nenhum é inferior a 5%, por conta disso adotamos um  α de 10%, o que influenciou o resultado das nossas regressões.

```python
X = dados_completos[['inbound exp 2012  (percentage)',
       'inbound exp 2018  (percentage)','outbound exp 2012  (percentage)',
       'outbound exp 2018  (percentage)', 'Bare area', 'Cropland',
       'Tree cover', 'Grassland', 'Shrubland', 'Sparse vegetation',
       'Artificial surfaces', 'Inland water', 'Wetland']]
Y = dados_completos['int inbound 2018']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)


results = regress(X_train,Y_train)
results.summary()
```

#### Gráfico de resíduos

```python
stats.probplot(results.resid,dist="norm",plot=plt);
```

```python
X_const= sm.add_constant(X_test)
Y_pred= results.predict(X_const)
```

```python
erro_quadratico_regressao = mean_squared_error(Y_test,Y_pred)
print('O erro da regressão linear foi de ',erro_quadratico_regressao**(1/2))
```

### Decision tree regressor

```python
tree_regressor = DecisionTreeRegressor(random_state = 0)
tree_regressor.fit(X_train, Y_train)
```

Plotamos esse gráfico para averiguar quais variáveis estavam influenciando mais na decision tree, que foram as categorias Shrubland (área arbustiva) e Bare Area (planice).

```python
tree_importance = tree_regressor.feature_importances_
plt.bar(X_train.columns,tree_importance)
plt.xticks(rotation=90)
```

```python
Y_pred_tree = tree_regressor.predict(X_test)
erro_quadratico_tree = mean_squared_error(Y_test,Y_pred_tree)
print('O erro da decision tree foi de ',erro_quadratico_tree**(1/2))
```

### Random Forest

```python
regressor = RandomForestRegressor(n_estimators = 200, random_state = 0,n_jobs=-1) 
regressor.fit(X_train, Y_train)
```

Fizemos novamente a plotagem de um gráfico para ver a influencia das variáveis no cálculo do número de turistas e dessa vez, as que mais levadas em conta foram novamente a de shrubland (área arbustiva) e a de sparce vegetation (Vegetação escassa), logo seguida por bare area (planice)

```python
importance = regressor.feature_importances_
plt.bar(X_train.columns,importance)
plt.xticks(rotation=90)
```

```python
Y_pred2=regressor.predict(X_test)
Erro_quadratico_forest = mean_squared_error(Y_test,Y_pred2)
print('O erro da random forest foi de ',Erro_quadratico_forest**(1/2))
```

## Conclusão

A partir do nosso projeto podemos concluir que a regressão que se mostrou melhor foi a Decision Tree, quando realizamos uma pesquisa sobre os diferentes tipos de métodos de análise de dados esperávamos que a Random Forest se mostrasse mais eficiente por se tratar de uma evolução da regressão Decision Tree, contudo nossa base de dados não facilitou o processo pois como foi apontado anteriormente existiam variáveis com correlação muito alta e as unidades numéricas apresentavam certa discrepância de unidade.

Por isso optamos por gerar um gráfico de feature importance visando observar quais as principais variáveis que estavam sendo escolhidas por cada método de regressão e que consequentemente acabaram afetando na pureza da nossa amostra. O que pode ser comum dependendo da base de dados analisada, que pode se adaptar melhor a um tipo de método do que outro, que foi o que aconteceu em nosso projeto.
